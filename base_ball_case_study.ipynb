# Modelling

Finding best random state

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
from sklearn.linear_model import LinearRegression
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state=i)
    mod = LinearRegression()
    mod.fit(x_train,y_train)
    pred = mod.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Maximum r2 score is ",maxAccu,"on Random_state ",maxRS)
```

I am getting the maximum R2 score as 92% on the random state 50

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30,random_state=maxRS)
```

# Regression Algorithm

```python
from sklearn.linear_model import LinearRegression, Lasso,Ridge
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn .neighbors import KNeighborsRegressor as KNN
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn import metrics 
```

# Linear Reggresion

```python
#checking r2 score for linear regression
LR = LinearRegression()
LR.fit(x_train,y_train)
 
#prediction
predLR = LR.predict(x_test)
print('R2_score:',r2_score(y_test,predLR))

#MAE
print('MAE:',metrics.mean_absolute_error(y_test,predLR))

#MSE
print('MSE:',metrics.mean_squared_error(y_test,predLR))

#RMSE
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,predLR)))

```

R2 score value is 92.91%

# Ridge Regressor


```python
ridge=Ridge(alpha=10.26,solver='sag')
ridge.fit(x_train,y_train)
ridge.score(x_train,y_train)
pred_ridge = ridge.predict(x_test)
print('R2_Score:',r2_score(y_test,pred_ridge))
print('MAE:',metrics.mean_squared_error(y_test,pred_ridge))
print('MSE:',metrics.mean_absolute_error(y_test,pred_ridge))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,pred_ridge)))
```

The R2 score for ridge regressor is 82.24%

# Lasso Regressor

```python
lasso=Lasso(alpha=1,random_state=0)
lasso.fit(x_train,y_train)
lasso.score(x_train,y_train)
pred_lasso=lasso.predict(x_test)
print('R2_Score:',r2_score(y_test,pred_lasso))
print('MAE:',metrics.mean_squared_error(y_test,pred_ridge))
print('MSE:',metrics.mean_absolute_error(y_test,pred_ridge))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,pred_ridge)))

```

# R2score using Lasso Regressor 86.42%

```python
# Checking R2 score for Stochastic Gradient Descent Regressor
SGD=SGDRegressor()
SGD.fit(x_train,y_train)

# prediction
predSGD=SGD.predict(x_test)
print('R2_Score:',r2_score(y_test,predSGD))
print('MAE:',metrics.mean_absolute_error(y_test, predSGD))
print('MSE:',metrics.mean_squared_error(y_test, predSGD))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, predSGD)))
```

```python
#checking r2 score for Stochastic Gradient Descent Regressor
SGD = SGDRegressor()
SGD.fit(x_train,y_train)
#prediction
predSGD=SGD.predict(x_test)
print('R2_Score:',r2_score(y_test,predSGD))
print('MAE:',metrics.mean_squared_error(y_test,pred_ridge))
print('MSE:',metrics.mean_absolute_error(y_test,pred_ridge))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_test,pred_ridge)))

```
